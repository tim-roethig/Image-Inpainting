```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import skimage
import os

def load_image_as_np_array(path):
    img = Image.open(path)
    return np.array(img)

def show_image_from_array(array):
    plt.imshow(array)
    
def create_mask(size_x, size_y, n=1, blocked_abs=None, blocked_rel=0.05):
    mask = np.ones((size_x, size_y, 4), dtype=int)
    x = np.random.randint(0, size_x)
    y = np.random.randint(0, size_y)
    range_x = np.random.randint(30, 100)
    range_y = np.random.randint(40, 100)
    for i in range(abs(x - range_x), min(x + range_x, size_x - 1)):
        for j in range(abs(y - range_y), min(y + range_y, size_y - 1)):
            mask[i][j] = np.array([0, 0, 0, 0])
    return mask
```

```python
img = load_image_as_np_array('test1.png')
mask = create_mask(img.shape[0], img.shape[1])
mx = np.multiply(img, mask)
show_image_from_array(mx)
```
